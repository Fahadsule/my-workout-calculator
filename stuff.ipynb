```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Your data
x = np.arange(0, 600.5, 0.5)  # Weight: 0 to 600
z = np.arange(0, 13)          # Reps: 0 to 12

# Create meshgrid
X, Z = np.meshgrid(x, z)
Y = X * (1 + (Z / 30))

# Create plot
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(X, Z, Y, cmap='viridis', alpha=0.8, edgecolor='none')

# Labels
ax.set_xlabel('Weight (lbs/kg)', fontsize=12, labelpad=10)
ax.set_ylabel('Reps', fontsize=12, labelpad=10)
ax.set_zlabel('1RM Prediction', fontsize=12, labelpad=10)
ax.set_title('1RM = Weight  (1 + Reps/30)', fontsize=16, pad=20)

# Colorbar
fig.colorbar(surf, shrink=0.5, aspect=10, label='1RM Value')

# Adjust view angle
ax.view_init(elev=25, azim=45)

plt.tight_layout()
plt.show()
```
